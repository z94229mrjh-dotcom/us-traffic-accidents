## Check for Low Variance / High Cardinality Columns

```python
num_cols = df.select_dtypes(include=[np.number]).columns

low_variance = []

for col in num_cols:
    col_var = df[col].var()
    top_freq = df[col].value_counts(normalize=True, dropna=False).iloc[0]

    low_variance.append({
        "column": col,
        "variance": col_var,
        "dominant_value_pct": top_freq
    })

low_variance_df = pd.DataFrame(low_variance).sort_values("variance")

low_variance_df
```

```python
cat_cols = df.select_dtypes(include=["object", "category"]).columns

high_cardinality = []

for col in cat_cols:
    nunique = df[col].nunique(dropna=False)
    total = len(df)
    ratio = nunique / total
    
    high_cardinality.append({
        "column": col,
        "n_unique": nunique,
        "cardinality_ratio": ratio
    })

high_card_df = pd.DataFrame(high_cardinality).sort_values("n_unique", ascending=False)

high_card_df
```

## Domain Validity Checks

```python
issues = {}

if {"start_lat", "start_lng"}.issubset(df.columns):
    lat_ok = df["start_lat"].between(-90, 90) | df["start_lat"].isna()
    lng_ok = df["start_lng"].between(-180, 180) | df["start_lng"].isna()
    issues["invalid_lat"] = (~lat_ok).sum()
    issues["invalid_lng"] = (~lng_ok).sum()

if "severity" in df.columns:
    sev_ok = df["severity"].isin([1, 2, 3, 4]) | df["severity"].isna()
    issues["invalid_severity"] = (~sev_ok).sum()

if {"start_time", "end_time"}.issubset(df.columns):
    time_order_ok = (df["end_time"] >= df["start_time"]) | df["end_time"].isna()
    issues["end_before_start"] = (~time_order_ok).sum()

if "distancemi" in df.columns:
    dist_ok = (df["distancemi"] >= 0) | df["distancemi"].isna()
    issues["negative_distance"] = (~dist_ok).sum()

issues
```

## Duplicate Checks

```python
dup_subset = ["id"] if "id" in df.columns else None

if dup_subset:
    n_dups = df.duplicated(subset=dup_subset).sum()
    print("Duplicates by ID:", n_dups)
    df_clean = df.drop_duplicates(subset=dup_subset)
else:
    n_dups = df.duplicated().sum()
    print("Full-row duplicates:", n_dups)
    df_clean = df.drop_duplicates()
```
